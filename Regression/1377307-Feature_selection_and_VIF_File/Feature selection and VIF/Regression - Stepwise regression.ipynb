# Iterative feature selection (Forward selection)

```python
feature_name =['AT', 'V', 'AP', 'RH', 'PE']
```

```python
import statsmodels.formula.api as smf
```

```python
data.corr()
```

```python
results = smf.ols("PE~AT",data=data).fit()
results.summary()
```

```python
results = smf.ols("PE~AT+AP",data=data).fit()
results.summary()
```

```python
results = smf.ols("PE~AT+AP+V",data=data).fit()
results.summary()
```

```python
results = smf.ols("PE~AT+AP+V+RH",data=data).fit()
results.summary()
```

```python
for col in ["AP","V","RH"]:
    print(f"results for formula = PE~ AT+ {col}")
    print("----------------------------------------")
    results = smf.ols(f"PE~ AT+ {col}",data=data).fit()
    print(results.summary(), end="\n\n")
```

# VIF  (Variance inflation factor)

```python
x_feature = data[['AT', 'V', 'AP', 'RH']]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = [variance_inflation_factor(x_feature.values,index)  for index, col in enumerate(x_feature.columns)]
```

```python
data_vif = pd.DataFrame(vif,columns=["vif"])
```

```python
data_vif["columns"]= ['AT', 'V', 'AP', 'RH']
```
